```sql
-- Test Service Type Based Amount Reporting
--Work Orders
SET spark.sql.parquet.datetimeRebaseModeInWrite = LEGACY;
SET spark.sql.parquet.datetimeRebaseModeInRead = LEGACY;

-- Step 1: Create the New Table if it does not exist using Delta format
CREATE TABLE IF NOT EXISTS DE_LH_SILVER.FleetData_WorkOrders_STB (
    No_ STRING,
    Posting_Date DATE,
    Starting_Date DATE,
    Finishing_Date DATE,
    Sell_to_Customer_No_ STRING,
    Sell_to_Customer_Name STRING,
    Bill_to_Customer_No_ STRING,
    Responsibility_Center STRING,
    Equipment_Object STRING,
    Key_Hours DECIMAL(38, 20),
    Pump_Hours DECIMAL(38, 20),
    Drive_Hours DECIMAL(38, 20),
    Service_Type STRING,
    CBilled_Line_Service_Types STRING,
    Parts_Used STRING,
    Parts_Used_Names STRING,
    Parts_Used_Manufacturer STRING,
    Parts_Used_Category STRING,
    Parts_Used_Full STRING,
    Fleet_Code STRING,
    Manufacturer_Code STRING,
    Equipment_Model STRING,
    Model_Year INT,
    Serial_No_ STRING,
    Parts DECIMAL(38, 20),
    Labor DECIMAL(38, 20),
    Subtotal DECIMAL(38, 20),
    Tax DECIMAL(38, 20),
    Amount_Including_Tax DECIMAL(38, 20)
) USING DELTA;

-- Step 2 & 3: Combine Data and Overwrite the Table
WITH Combined_Data AS (
    SELECT
        WOH.No_,
        WOH.Posting_Date,
        WOH.Starting_Date,
        WOH.Finishing_Date,
        WOH.Sell_to_Customer_No_,
        WOH.Sell_to_Customer_Name,
        WOH.Bill_to_Customer_No_,
        WOH.Responsibility_Center,
        WOH.Equipment_Object,
        WOH.M1 AS Key_Hours,
        WOH.M2 AS Pump_Hours,
        WOH.M3 AS Drive_Hours,
        WOH.Service_Type,
        CONCAT_WS(
            ', ',
            COLLECT_LIST(
                DISTINCT CASE
                    WHEN STB.Type = 'Customer Billed'
                         AND WOL.Amount_incl_Tax > 0
                    THEN WOL.Service_Type
                    ELSE NULL
                END
            )
        ) AS CBilled_Line_Service_Types,
        CONCAT(
            '',
            CONCAT_WS(', ', COLLECT_LIST(DISTINCT I.No_))
        ) AS Parts_Used,
        CONCAT(
            '',
            CONCAT_WS(', ', COLLECT_LIST(DISTINCT I.Description))
        ) AS Parts_Used_Names,
        CONCAT(
            '',
            CONCAT_WS(', ', COLLECT_LIST(DISTINCT IC.Description))
        ) AS Parts_Used_Category,
        CONCAT(
            '',
            CONCAT_WS(', ', COLLECT_LIST(DISTINCT I.Manufacturer_Code))
        ) AS Parts_Used_Manufacturer,
        CONCAT_WS(
            '; ',
            COLLECT_LIST(
            DISTINCT CONCAT(
                I.No_, ', ', 
                I.Description, ', ', 
                I.Manufacturer_Code, ', ', 
                IC.Description
                )
            )
        ) AS Parts_Used_Full,
        WOH.Fleet_Code,
        WOH.Manufacturer_Code,
        WOH.Equipment_Model,
        EO.Year AS Model_Year,
        WOH.Serial_No_,
        SUM(
            CASE
                WHEN STB.Type = 'Customer Billed'
                     AND (
                        (WOL.Type IN (1, 3, 6, 10, 11))
                        OR (WOL.Gen_Prod_Posting_Group = 'FREIGHT')
                    )
                THEN WOL.Amount
                ELSE 0
            END
        ) AS Parts,
        SUM(
            CASE 
                WHEN STB.Type = 'Customer Billed'
                     AND (
                        (WOL.Type IN (2) AND WOL.Gen_Prod_Posting_Group <> 'FREIGHT')
                        OR (
                            WOL.Type IN (5)
                            AND (JC.Code LIKE '%PM%' OR JC.Job_Code_Group LIKE '%PM%')
                        )
                    )
                THEN WOL.Amount
                ELSE 0
            END
        ) AS Labor,
        SUM(
            CASE
                WHEN STB.Type = 'Customer Billed'
                     AND (
                        (WOL.Type IN (1, 3, 6, 10, 11))
                        OR (WOL.Gen_Prod_Posting_Group = 'FREIGHT')
                    )
                THEN WOL.Amount
                ELSE 0
            END
        )
        +
        SUM(
            CASE 
                WHEN STB.Type = 'Customer Billed'
                     AND (
                        (WOL.Type IN (2) AND WOL.Gen_Prod_Posting_Group <> 'FREIGHT')
                        OR (
                            WOL.Type IN (5)
                            AND (JC.Code LIKE '%PM%' OR JC.Job_Code_Group LIKE '%PM%')
                        )
                    )
                THEN WOL.Amount
                ELSE 0
            END
        ) AS Subtotal,
        SUM(
            CASE
                WHEN STB.Type = 'Customer Billed'
                THEN WOL.Amount_incl_Tax
                ELSE 0
            END
        )
        - SUM(
            CASE
                WHEN STB.Type = 'Customer Billed'
                THEN WOL.Amount
                ELSE 0
            END
        ) AS Tax,
        SUM(
            CASE
                WHEN STB.Type = 'Customer Billed'
                THEN WOL.Amount_incl_Tax
                ELSE 0
            END
        ) AS Amount_Including_Tax
    FROM DE_LH_SILVER.Work_Order_Header AS WOH
    INNER JOIN DE_LH_SILVER.Work_Order_Line AS WOL
        ON WOH.No_ = WOL.Document_No_
    LEFT JOIN DE_LH_SILVER.Job_Code AS JC
        ON WOL.Job_Code = JC.Code
    LEFT JOIN DE_LH_BRONZE_ELC.Service_Type_Billing AS STB
        ON WOL.Service_Type = STB.Code
    LEFT JOIN DE_LH_SILVER.Item AS I
        ON WOL.No_ = I.No_
    LEFT JOIN DE_LH_SILVER.Item_Category AS IC
        ON I.Item_Category_Code = IC.Code
    LEFT JOIN DE_LH_SILVER.Equipment_Object AS EO
        ON WOH.Equipment_Object = EO.No_
    WHERE 
        WOH.Posting_Date BETWEEN DATE('2022-01-01') 
                            AND current_date()
        AND WOH.Posting_Status = 2 AND WOH.Document_Type = 1
    GROUP BY
        WOH.No_,
        WOH.Posting_Date,
        WOH.Starting_Date,
        WOH.Finishing_Date,
        WOH.Sell_to_Customer_No_,
        WOH.Sell_to_Customer_Name,
        WOH.Bill_to_Customer_No_,
        WOH.Responsibility_Center,
        WOH.Equipment_Object,
        WOH.M1,
        WOH.M2,
        WOH.M3,
        WOH.Service_Type,
        WOH.Fleet_Code,
        WOH.Manufacturer_Code,
        WOH.Equipment_Model,
        EO.Year,
        WOH.Serial_No_
)

-- Step 3: Overwrite the table with the combined results
INSERT OVERWRITE TABLE DE_LH_SILVER.FleetData_WorkOrders_STB
SELECT 
    No_,
    Posting_Date,
    Starting_Date,
    Finishing_Date,
    Sell_to_Customer_No_,
    Sell_to_Customer_Name,
    Bill_to_Customer_No_,
    Responsibility_Center,
    Equipment_Object,
    ROUND(Key_Hours, 2) AS Key_Hours,
    ROUND(Pump_Hours, 2) AS Pump_Hours,
    ROUND(Drive_Hours, 2) AS Drive_Hours,
    Service_Type,
    CBilled_Line_Service_Types,
    Parts_Used,
    Parts_Used_Names,
    Parts_Used_Manufacturer,
    Parts_Used_Category,
    Parts_Used_Full,
    Fleet_Code,
    Manufacturer_Code,
    Equipment_Model,
    Model_Year,
    Serial_No_,
    ROUND(Parts, 2) AS Parts,
    ROUND(Labor, 2) AS Labor,
    ROUND(Subtotal, 2) AS Subtotal,
    ROUND(Tax, 2) AS Tax,
    ROUND(Amount_Including_Tax, 2) AS Amount_Including_Tax
FROM Combined_Data;

-- Step 4: View the Results (optional)
SELECT * 
FROM DE_LH_SILVER.FleetData_WorkOrders_STB 
WHERE `No_` = 'S2663870';
```

# FleetData_Comments

```sql
SET spark.sql.parquet.datetimeRebaseModeInRead = LEGACY;
SET spark.sql.parquet.int96RebaseModeInRead    = LEGACY;

-- STEP 1: Create your final table if not exists
CREATE TABLE IF NOT EXISTS DE_LH_SILVER.FleetData_Comments (
  No_ STRING,
  Sell_to_Customer_No_ STRING,
  Equipment_Object STRING,
  Reason STRING,
  Comment STRING,
  Attachment STRING,
  Damage STRING,
  Tires STRING,
  Battery STRING
)
USING DELTA;

--------------------------------------------------------------------------------
-- STEP 2: Collect 'Reason' lines, sorted by (Doc_Line, Subline)
--------------------------------------------------------------------------------
WITH Rental_Comment_Line_Reason AS (
  SELECT
    RCL.Doc_No_ as Doc_No_,
    -- Build a struct with (Doc_Line, Subline, Comment)
    named_struct(
      'doc_line', RCL.Doc_Line,
      'Subline', RCL.Subline,
      'comment',  RCL.Comment
    ) as rowval
  FROM DE_LH_SILVER.Rental_Comment_Line AS RCL
  WHERE RCL.Doc_Line = 0
),
CTE_Reason AS (
  SELECT
    Doc_No_,
    UPPER(
      CONCAT_WS(
        ' ',
        -- Sort the array of structs in ascending order using the default comparator:
        TRANSFORM(
          SORT_ARRAY(COLLECT_LIST(rowval)),   -- no second argument to SORT_ARRAY
          x -> x.comment                      -- x is each struct; extract 'comment'
        )
      )
    ) AS Reason
  FROM Rental_Comment_Line_Reason
  GROUP BY Doc_No_
),

--------------------------------------------------------------------------------
-- STEP 3: Collect 'Comment' lines, sorted by (Doc_Line, Subline)
--------------------------------------------------------------------------------
Rental_Comment_Line_Comment AS (
  SELECT
    RCL.Doc_No_ as Doc_No_,
    named_struct(
      'doc_line', RCL.Doc_Line,
      'Subline', RCL.Subline,
      'comment',  RCL.Comment
    ) as rowval
  FROM DE_LH_SILVER.Rental_Comment_Line AS RCL
  WHERE RCL.Doc_Line <> 0
),
CTE_Comment AS (
  SELECT
    Doc_No_,
    UPPER(
      CONCAT_WS(
        ' ',
        TRANSFORM(
          SORT_ARRAY(COLLECT_LIST(rowval)),
          x -> x.comment
        )
      )
    ) AS Comment
  FROM Rental_Comment_Line_Comment
  GROUP BY Doc_No_
),

--------------------------------------------------------------------------------
-- STEP 4: JOIN with Work_Order_Header
--------------------------------------------------------------------------------
CTE_WorkOrder AS (
  SELECT
    WOH.No_,
    WOH.Sell_to_Customer_No_,
    WOH.Equipment_Object,
    COALESCE(R.Reason,  '') AS Reason,
    COALESCE(C.Comment, '') AS Comment
  FROM DE_LH_SILVER.Work_Order_Header AS WOH
  LEFT JOIN CTE_Reason  R ON WOH.No_ = R.Doc_No_
  LEFT JOIN CTE_Comment C ON WOH.No_ = C.Doc_No_
  WHERE YEAR(WOH.Posting_Date) IN (
    YEAR(CURRENT_DATE()) - 1,
    YEAR(CURRENT_DATE())
  )
)

--------------------------------------------------------------------------------
-- STEP 5: Insert/Overwrite your final table
--------------------------------------------------------------------------------
/* ——— STEP 5: replace existing INSERT block with this MERGE ——— */
MERGE INTO DE_LH_SILVER.FleetData_Comments AS tgt
USING (
    /* one row per work-order from CTE_WorkOrder */
    SELECT
        No_,
        Sell_to_Customer_No_,
        Equipment_Object,
        Reason,
        Comment,

        /* Attachment flag */
        CASE
            WHEN INSTR(Comment,'ATTACH')  > 0
              OR INSTR(Reason, 'ATTACH')  > 0
            THEN 'Yes' ELSE '' END              AS Attachment,

        /* Damage flag  (add more keywords here if needed) */
        CASE
            WHEN INSTR(Comment,'AVOID')   > 0 OR INSTR(Reason,'AVOID')   > 0
              OR INSTR(Comment,'ABUSE')   > 0 OR INSTR(Reason,'ABUSE')   > 0
              OR INSTR(Comment,'DAMAGE')  > 0 OR INSTR(Reason,'DAMAGE')  > 0
            THEN 'Yes' ELSE '' END              AS Damage,

        /* Tires flag */
        CASE
            WHEN INSTR(Comment,'TIRE')    > 0 OR INSTR(Reason,'TIRE')    > 0
            THEN 'Yes' ELSE '' END              AS Tires,

        /* Battery flag */
        CASE
            WHEN INSTR(Comment,'BATTERY')   > 0 OR INSTR(Comment,'BATTERIES') > 0
              OR INSTR(Reason, 'BATTERY')   > 0 OR INSTR(Reason, 'BATTERIES') > 0
            THEN 'Yes' ELSE '' END              AS Battery
    FROM   CTE_WorkOrder
    WHERE  Reason <> '' OR Comment <> ''
) src
ON  tgt.No_ = src.No_

WHEN MATCHED THEN UPDATE SET
    tgt.Sell_to_Customer_No_ = src.Sell_to_Customer_No_,
    tgt.Equipment_Object     = src.Equipment_Object,
    tgt.Reason               = src.Reason,
    tgt.Comment              = src.Comment,
    tgt.Attachment           = src.Attachment,
    tgt.Damage               = src.Damage,
    tgt.Tires                = src.Tires,
    tgt.Battery              = src.Battery

WHEN NOT MATCHED THEN INSERT (
      No_, Sell_to_Customer_No_, Equipment_Object,
      Reason, Comment, Attachment, Damage, Tires, Battery
) VALUES (
      src.No_, src.Sell_to_Customer_No_, src.Equipment_Object,
      src.Reason, src.Comment, src.Attachment, src.Damage, src.Tires, src.Battery
);

-- (Optional) Preview results
SELECT * 
FROM DE_LH_SILVER.FleetData_Comments;
```

```sql
%%sql
--Maintenance Contracts
-- Step 1: Create the New Table if it does not exist using Delta format
CREATE TABLE IF NOT EXISTS DE_LH_SILVER.FleetData_MaintenanceContracts (
    No_ STRING,
    Contract_Type STRING,
    Sell_to_Customer_No_ STRING,
    Bill_to_Customer_No_ STRING,
    Responsibility_Center STRING,
    Equipment_Object STRING,
    Serial_No_ STRING,
    Equipment_Model STRING,
    Manufacturer_Code STRING,
    Contract_Status STRING,
    Object_Status STRING,
    Fleet_Code STRING,
    Document_No_ STRING,
    Posting_Date DATE,
    Description_2 STRING,
    ELC_Doc_Type INT,
    Amount DECIMAL(38, 20)
) USING DELTA;

-- Step 2 & 3: Combine Data and Overwrite the Table
WITH Combined_Data AS (
    SELECT
        MCH.No_,
        MCH.Contract_Type,
        MCH.Sell_to_Customer_No_,
        MCH.Bill_to_Customer_No_,
        MCH.Responsibility_Center,
        MCL.Equipment_Object,
        EO.Serial_No_,
        EO.Equipment_Model,
        EO.Manufacturer_Code,
        MCH.Status AS Contract_Status,
        MCL.Status AS Object_Status,
        MCL.Fleet_Code,
        SIL.Document_No_,
        SIL.Posting_Date,
        SIL.Description_2,
        SIL.ELC_Doc_Type,
        SIL.Amount
    FROM DE_LH_SILVER.Maintenance_Contract_Header AS MCH
    JOIN DE_LH_SILVER.Maintenance_Contract_Line AS MCL ON MCH.No_ = MCL.Document_No_
    LEFT JOIN DE_LH_SILVER.Sales_Invoice_Line AS SIL ON MCH.No_ = SIL.Contract_No_ AND MCL.Equipment_Object = SIL.Equipment_Object
    LEFT JOIN DE_LH_SILVER.Equipment_Object AS EO ON MCL.Equipment_Object = EO.No_
    WHERE YEAR(SIL.Posting_Date) IN (YEAR(current_date()) - 1, YEAR(current_date()))
        AND SIL.ELC_Doc_Type = 6
)
INSERT OVERWRITE TABLE DE_LH_SILVER.FleetData_MaintenanceContracts
SELECT 
    No_,
    Contract_Type,
    Sell_to_Customer_No_,
    Bill_to_Customer_No_,
    Responsibility_Center,
    Equipment_Object,
    Serial_No_,
    Equipment_Model,
    Manufacturer_Code,
    Contract_Status,
    Object_Status,
    Fleet_Code,
    Document_No_,
    Posting_Date,
    Description_2,
    ELC_Doc_Type,
    ROUND(Amount, 2) AS Amount
FROM Combined_Data;

-- Step 4: View the Results (optional)
SELECT * FROM DE_LH_SILVER.FleetData_MaintenanceContracts;
```

# FleetData_Credits

```sql
%%sql
--Credits
-- Step 1: Create the New Table if it does not exist using Delta format
CREATE TABLE IF NOT EXISTS DE_LH_SILVER.FleetData_Credits (
    No_ STRING,
    Posting_Date DATE,
    Sell_to_Customer_No_ STRING,
    Equipment_Object STRING,
    Parts_Credits DECIMAL(38, 20),
    Labor_Credits DECIMAL(38, 20),
    Credits_Subtotal DECIMAL(38, 20),
    Tax_Credits DECIMAL(38, 20),
    Total_Credits_Including_Tax DECIMAL(38, 20),
    Credit_Date DATE
) USING DELTA;

-- Step 2 & 3: Combine Data and Overwrite the Table
WITH Combined_Data AS (
    SELECT
        WOH.No_,
        WOH.Posting_Date,
        WOH.Sell_to_Customer_No_,
        WOH.Equipment_Object,
        SUM(CASE WHEN WOL.Type = 1 OR WOL.Gen_Prod_Posting_Group = 'FREIGHT' THEN SCML.Amount * -1 ELSE 0 END) AS Parts_Credits,
        SUM(CASE WHEN WOL.Type IN (2, 3) AND WOL.Gen_Prod_Posting_Group <> 'FREIGHT' THEN SCML.Amount * -1 ELSE 0 END) AS Labor_Credits,
        SUM(CASE WHEN WOL.Type = 1 THEN SCML.Amount * -1 ELSE 0 END) + 
        SUM(CASE WHEN WOL.Type IN (2, 3) THEN SCML.Amount * -1 ELSE 0 END) AS Credits_Subtotal,
        (SUM(SCML.Amount_Including_VAT) - SUM(SCML.Amount)) * -1 AS Tax_Credits,
        SUM(SCML.Amount_Including_VAT) * -1 AS Total_Credits_Including_Tax,
        MAX(SCMH.Posting_Date) AS Credit_Date
    FROM 
        DE_LH_SILVER.Work_Order_Header AS WOH
    INNER JOIN 
        DE_LH_SILVER.Work_Order_Line AS WOL ON WOH.No_ = WOL.Document_No_
    INNER JOIN 
        DE_LH_SILVER.Sales_Invoice_Line AS SIL ON WOL.Document_No_ = SIL.ELC_Document_No_ AND WOL.Line_No_ = SIL.Work_Order_Line_No_
    INNER JOIN
        DE_LH_SILVER.Sales_Credit_Memo_Line AS SCML ON WOL.Document_No_ = SCML.Work_Order_No_ AND WOL.Line_No_ = SCML.Work_Order_Line_No_
    INNER JOIN
        DE_LH_SILVER.Sales_Credit_Memo_Header AS SCMH ON SCML.Document_No_ = SCMH.No_
    WHERE
        YEAR(WOH.Posting_Date) IN (YEAR(current_date()) - 1, YEAR(current_date()))
        AND WOH.Posting_Status = 2
    GROUP BY
        WOH.No_,
        WOH.Posting_Date,
        WOH.Sell_to_Customer_No_,
        WOH.Equipment_Object
)
INSERT OVERWRITE TABLE DE_LH_SILVER.FleetData_Credits
SELECT 
    No_,
    Posting_Date,
    Sell_to_Customer_No_,
    Equipment_Object,
    ROUND(Parts_Credits, 2) AS Parts_Credits,
    ROUND(Labor_Credits, 2) AS Labor_Credits,
    ROUND(Credits_Subtotal, 2) AS Credits_Subtotal,
    ROUND(Tax_Credits, 2) AS Tax_Credits,
    ROUND(Total_Credits_Including_Tax, 2) AS Total_Credits_Including_Tax,
    Credit_Date
FROM Combined_Data;

-- Step 4: View the Results (optional)
SELECT * FROM DE_LH_SILVER.FleetData_Credits;
```

# FleetData_HoursCheck

```sql
%%sql
--Hours Check
-- Step 1: Create the New Table if it does not exist using Delta format
CREATE TABLE IF NOT EXISTS DE_LH_SILVER.FleetData_HoursCheck (
    Equipment_Object STRING,
    Posting_Date DATE,
    Key_Hours DECIMAL(38, 20),
    Pump_Hours DECIMAL(38, 20),
    Drive_Hours DECIMAL(38, 20),
    Hour_Check STRING
) USING DELTA;

-- Step 2 & 3: Combine Data and Overwrite the Table
WITH RankedData AS (
    SELECT
        WOH.Equipment_Object,
        WOH.Posting_Date,
        WOH.M1 AS Key_Hours,
        WOH.M2 AS Pump_Hours,
        WOH.M3 AS Drive_Hours,
        ROW_NUMBER() OVER (PARTITION BY WOH.Equipment_Object ORDER BY WOH.Posting_Date DESC) AS RowNum,
        LAG(WOH.M1) OVER (PARTITION BY WOH.Equipment_Object ORDER BY WOH.Posting_Date) AS PrevKeyHours,
        LAG(WOH.M2) OVER (PARTITION BY WOH.Equipment_Object ORDER BY WOH.Posting_Date) AS PrevPumpHours,
        LAG(WOH.M3) OVER (PARTITION BY WOH.Equipment_Object ORDER BY WOH.Posting_Date) AS PrevDriveHours
    FROM DE_LH_SILVER.Work_Order_Header AS WOH
    WHERE
        WOH.Posting_Status = 2
        AND WOH.Posting_Date <> '1753-01-01'
        AND (WOH.Equipment_Object LIKE 'E%' OR WOH.Equipment_Object LIKE 'H%')
)
INSERT OVERWRITE TABLE DE_LH_SILVER.FleetData_HoursCheck
SELECT
    Equipment_Object,
    Posting_Date,
    Key_Hours,
    Pump_Hours,
    Drive_Hours,
    CASE 
        WHEN Key_Hours < PrevKeyHours THEN 'Key Hours'
        ELSE ''
    END +
    CASE 
        WHEN Pump_Hours < PrevPumpHours THEN 
            CASE 
                WHEN Key_Hours < PrevKeyHours THEN ', Pump Hours' 
                ELSE 'Pump Hours' 
            END
        ELSE ''
    END +
    CASE 
        WHEN Drive_Hours < PrevDriveHours THEN 
            CASE 
                WHEN Key_Hours < PrevKeyHours OR Pump_Hours < PrevPumpHours THEN ', Drive Hours' 
                ELSE 'Drive Hours' 
            END
        ELSE ''
    END AS Hour_Check
FROM RankedData
WHERE RowNum <= 10
ORDER BY Equipment_Object, Posting_Date DESC;

-- Step 4: View the Results (optional)
SELECT * FROM DE_LH_SILVER.FleetData_HoursCheck WHERE Equipment_Object = 'H8FGC70U'
```
